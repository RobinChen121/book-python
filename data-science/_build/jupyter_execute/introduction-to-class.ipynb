# Python Object Oriented Programming*
<hr>

Python is a programming language that supports object-oriented programming (OOP). Object-oriented programming is a paradigm that encapsulates data and the functions that operate on it within objects. By organizing code through classes and objects, **it makes programs clearer, easier to maintain, and more extensible and reusable**, making it particularly well-suited for developing large and complex systems.

## Class, Object, Attribute, Method
<hr>

In object-oriented programming, a class is a template or blueprint for objects, describing the attributes (variables) and behaviors (methods) of a certain type of entity. An object, on the other hand, is an instance of a class, concretely embodying the class’s attributes and behaviors. For example, if we define "Employee" as a class, it can include attributes such as name, salary, years of service, and position; a specific employee, like Zhang San or Li Si, would be an instance of this class. To use an analogy, a class is like a “cake mold,” and an object is the “cake” made from that mold.

Python uses `class` to define a class, and within the class further defines class attributes (data) and methods (functions). The syntax is as follows:

In the following example, we define an Employee class.

`````python
class Employee:
    # Class attributes
    department = "business school"  # Regular attribute
    __age = 20  # Variables prefixed with double underscores are private class attributes

    # Class methods
    def get_age(self):
        return self.__age

    def say_hello(self):
        print("Hello, everyone!")


john = Employee()  # Instantiate a class
john.say_hello()  # Access a method of the object
print("My department is %s." % john.department)  # Access an attribute of the object
`````

In the example above, we defined an `Employee` class with two attributes: a regular attribute and a private attribute. They can both have default values inside the defintion. By instantiating the class, we can create an object and access some of the attributes and methods defined in the class through the instance.

```{note}
Variables prefixed with double underscores (`__`) are treated as private attributes of a class and cannot be accessed directly from outside the class. However, we can define a method within the class to access the private attribute and then retrieve its value by calling that method.
```

`````python
print("age is", john.get_age())
`````

If you try to directly access the private attribute __age from the object john (for example below), the program will raise an error.

````{code-block} python3
print(john.__age)
````

````{note}
- Variables or functions prefixed with double underscores (`__`) are considered private attributes or methods of a class and cannot be accessed directly from outside the class.

- Variables or functions prefixed with a single underscore (`_`) are considered protected attributes or methods of a class. They can be accessed from outside the class, but doing so is not recommended.
````

In Python, you can dynamically add attributes to classes or instances, which means you can assign a value to an attribute that was not originally defined in the class.

`````python
john._weight = 70
print(john._weight)
`````

## Initialize a class `__init__`
<hr>

A class has a special method called `__init__()` that is automatically invoked when an instance of the class is created, and is used to initialize the object. For example:

`````python
class Employee:
    # Class attributes
    department = "business school"  # Regular attribute
    __age = 20  # Variables with double underscores are private class attributes

    # Class initializer
    def __init__(self, name, age, department):
        self.name = name
        self.__age = age
        self.department = department

    # Class methods
    def get_age(self):
        return self.__age

    def say_hello(self):
        print("Hello, everyone! My name is %s." % self.name)


john = Employee("John", 25, "engineering school")  # Instantiate the class; __init__() is called automatically
john.say_hello()  # Call a method of the class
print("My department is %s." % john.department)  # Access an attribute of the class
`````

In the example above, the `__init__` method initializes multiple attributes of the class, and any default attribute values are overwritten by the values passed as parameters to the `__init__` method.

````{note}
A key difference between class methods and regular functions is that the first parameter of class methods is always `self`, which represents the instance of the class.
````

## Class Inheritance and Method Overriding
<hr>

Python supports class inheritance, meaning that one class can `inherit` all the attributes and methods of another class. It can also add new attributes and methods, or `override` methods of the parent class.

For example, we can create a class 'NewEmployee' that inherits from the 'Employ' class.

`````python
class NewEmployee(Employee):  # The class inside the parentheses is the parent class being inherited
    under_probation = True  # Attribute specific to the subclass

    def __init__(self, under_probation, *args, **kwargs):
        super().__init__(*args, **kwargs)  # Call the parent class's initializer
        self.under_probation = under_probation

    # Override a method from the parent class
    def say_hello(self):
        print("Hi! My name is %s." % self.name)

    # Call a method from the parent class
    def age_range(self):
        if super().get_age() >= 30:  # Call the parent class's method
            print("I am over 30s")
        else:
            print("I am under 30s")

    # Method specific to the subclass
    def probabtion_info(self):
        if self.under_probation:
            print("I am under probation.")
        else:
            print("I have passed probation.")


tom = NewEmployee(
    under_probation=False, name="John", age=25, department="engineering school"
)
tom.say_hello()
tom.probabtion_info()
`````

In the above example, `super()` is used to call the parent class’s `__init__` and `get_age` methods, and the parent class’s `say_hello` method is overridden.

Python also supports multiple inheritance, meaning that a subclass can inherit from multiple parent classes by placing multiple class names inside the parentheses of the subclass. For example, we can define a 'NewAcademicEmployee' class that inherits from both the 'Employee' and 'NewEmployee' classes.

`````python
class NewAcademicEmployee(NewEmployee, Employee):
    def say_hello(self):
        print("Hello, I am a new academic empolyee.")


# When instantiating a subclass, 
# the parameters of its initializer include all the parameters 
# required by the initializers of its parent classes.
lily = NewAcademicEmployee(True, "Lily", 22, "art school")
lily.say_hello()
`````

In the example above, when initializing the subclass, all the parameters required by the initializers of the parent classes need to be passed.

```{note}
If a subclass and its parent class have methods with the same name, and you call that method on an instance of the subclass, Python will prioritize the subclass’s method and will not call the parent class’s method.
```

`````python
class Parent:
    def show(self):
        print("Parent method")


class Child(Parent):
    def show(self):  # override the parent's method
        print("Child method")


obj = Child()
obj.show()  # output: Child's method
`````

## *Special (or “magic”) methods of a class
<hr>

Methods named with double underscores at both ends are special or “magic” methods of a class, used to implement specific functionalities.

The table below summarizes some commonly used special methods.

| Magic Method | Description |
| :--- | :--- |
| `__init__` | Called automatically when an object is created; used to initialize the instance. |
| `__str__` | User-oriented; defines the string representation when the object is printed. |
| `__repr__` | Developer-oriented; used for debugging to show the official object representation. |
| `__getitem__` | Gets an element, allowing the object to be accessed like an array or dictionary. |
| `__setitem__` | Sets an element, allowing the object to be modified like an array or dictionary. |
| `__add__` | Implements the addition operation (`+`) for the object. |
| `__sub__` | Implements the subtraction operation (`-`) for the object. |
| `__mul__` | Implements the multiplication operation (`*`) for the object. |
| `__truediv__` | Implements the division operation (`/`) for the object. |
| `__eq__` | Defines the rules for equality comparison (`==`) between objects. |
| `__hash__` | Defines the hash value, allowing the object to be stored in a `set` or used as a `dict` key. |
| `__class__` | Returns the reference to the class that the object belongs to. |
| `__call__` | Allows the object instance to be called like a function. |
| `__len__` | Defines the behavior for the `len()` function; returns the size of the object. |

The following example shows the usages of `__repr__` and `__str__`.

`````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __repr__(self):
        return f"Person('{self.name}', {self.age})"

    def __str__(self):
        return f"{self.name}, {self.age} years old"


p = Person("Bob", 30)
print(repr(p))  # call __repr__
print(p)  # __str__ is called by default; if __str__ is not defined, __repr__ will be called as a fallback.
`````

If the `__repr__` or `__str__` methods are not defined, printing the object will display its memory address.

`````python
class Dog:
    pass


d = Dog()
print(d)  # there are no method definitions for  __str__  or __repr__ and print will output the memory address
print(d.__class__)  # show the reference of the class
`````

````{note}
`__main__` refers to the currently running Python file (the file with a .py extension).
````

The following example demonstrates how to use the `__getitem__` method to access an object just like an array.

`````python
class MyList:
    def __init__(self, data):
        self.data = data

    def __getitem__(self, index):
        return self.data[index]  


my_list = MyList([10, 20, 30, 40, 50])
print(my_list[1:3])  
`````

The following example demonstrates how to add two classes together.

`````python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"


v1 = Vector(2, 3)
v2 = Vector(4, 5)
print(v1 + v2) 
`````

The following example shows how to compare two classes.

`````python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __eq__(self, other):
        if isinstance(other, Person):
            return self.name == other.name and self.age == other.age
        return False 


p1 = Person("Alice", 25)
p2 = Person("Alice", 25)
p3 = Person("Bob", 30)

print(p1 == p2) 
print(p1 == p3) 
`````

The following example illustrates how to make an object callable like a function.

`````python
class MyClass:
    def __call__(self, *args, **kwargs):
        print("Instance called!")
        print("Arguments:", args, kwargs)


obj = MyClass()
obj(1, 2, a=3)
`````

## *Special (or "magic") attribute of a class 
<hr>

In Python, special attributes of a class are also known as “magic attributes” or “dunder” (double underscore) attributes/methods. They typically refer to special attributes or methods that begin and end with **double underscores**. These attributes and methods are built into the Python language and are used to implement specific behaviors and functionalities.

For example:

- `__doc__`: Stores the docstring of a module, class, or function.

- `__name__`: Records the name of a module or class.

- `__dict__`: Holds the attribute dictionary of an object, containing all of its writable attributes.

`````python
class Example:
    """This is a sample class demonstrating magic attributes."""

    def __init__(self, value):
        self.value = value

    def __str__(self):
        return f"Example object with value: {self.value}"

    def __repr__(self):
        return f"Example({self.value!r})"

# Create an instance
ex = Example(42)

# Using magic attributes
print(ex.__doc__)   # Output: This is a sample class demonstrating magic attributes.
print(str(ex))      # Output: Example object with value: 42
print(repr(ex))     # Output: Example(42)
print(ex.__dict__)  # Output the instance's attribute values as a dictionary
`````

`````python
MyClass.__name__  # return the name of a class
`````

`````python
import numpy as np

np.__name__  # return the name of a module
`````

`````python
# myscript.py
def hello():
    print("Hello, world!")


if __name__ == "__main__":
    print("This script is running directly")
    hello()
else:
    print("This script is imported as a module")
`````

## Attribute related methods`setattr()`, `getattr()`
<hr>

For modifying and accessing attribute values in a class, Python provides two built-in functions that are quite useful: `setattr()` and `getattr()`.

- `setattr(object, name, value)`: sets the value of the attribute name on object to value. If the attribute does not exist, it is created and assigned the given value.
- `getattr(object, name)`: returns the value of the attribute name on object. If the attribute is not found, raises AttributeError.
- `getattr(object, name, default)`: returns the value of the attribute name on object. If the attribute is not found, returns the provided default value instead of raising an exception.

Among them, the attribute parameters in parentheses are represented as strings.

`````python
class ExampleClass:
    
    def __init__(self, name, age):
        self.name = name
        self.age = age

setattr(ExampleClass, 'name', 10)
print(getattr(ExampleClass, 'name'))
setattr(ExampleClass, 'gender', 'male')  # can not find the attribute 'gender', then it will create this attribute with given value
print(getattr(ExampleClass, 'gender'))

p = ExampleClass('tom', '40')
print(getattr(p, 'grade', 75)) # can not find the velue for the attribute 'grade', then it output the given default value
print(p.gender)
`````

## *Class method and static method
<hr>

In the examples above, the first parameter of a method in a class is always self. Such methods are called `instance methods` and can only be called by an object.

In addition to `instance methods`, there are also `class methods` and `static methods`.

- Static methods: Methods decorated with `@staticmethod` that do not take a self parameter are called static methods. A class’s static method can have no parameters and can be called directly using the class name.

- Class methods: These methods have a `cls` parameter by default, can be called by both the class and its instances, and must be decorated with @classmethod.

`````python
class MyClass:
    @staticmethod  # Names starting with @ are decorators, used to indicate special functions
    def fun():
        print("Static method")

    @classmethod
    def a(cls):  # The first parameter is cls
        print("Class method")

    # Instance method
    def b(self):  # The first parameter is self
        print("Instance method")


MyClass.fun()  # Static methods can be called directly by the class name
MyClass.a()    # Class methods can be called directly by the class name

case = MyClass()
case.fun()  # Static methods can also be called by an instance
case.a()    # Class methods can also be called by an instance
case.b()    # Instance methods can only be called by an instance
`````

## dateclass
<hr>

`dataclass` is a decorator introduced in Python 3.7 that automatically generates methods like `__init__`, `__repr__`, and `__eq__`, making the definition of data classes more concise.

One example:

`````python
from dataclasses import dataclass


@dataclass
class Person:
    name: str  # declare name as string type
    age: int  # declare age as an int type
`````

The above codes are equivalent to:

`````python
class Person:
    def __init__(self, name: str, age: int):  # A colon followed by a data type can be used to declare the type of a variable.
        self.name = name
        self.age = age

    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"
`````

It can be seen that one difference between a dataclass and a regular class is that the data types of attributes must be declared in the definition, while the amount of code is significantly reduced.
